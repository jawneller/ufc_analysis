```python
# Create dfs for each csv in the directory, named the same
for f in os.listdir():
    if f.endswith('.csv'):
        df_name = os.path.splitext(f)[0]
        globals()[df_name] = pd.read_csv(f)
        print(f"{df_name} contains {len(globals()[df_name].index)} rows")
```
